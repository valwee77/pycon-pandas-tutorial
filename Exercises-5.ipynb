```python
cast = pd.read_csv('data/cast.csv')
cast.head()
```

```python
release_dates = pd.read_csv(
    'data/release_dates.csv',
    parse_dates=['date'],
    infer_datetime_format=True,
)
release_dates.head()
```

### Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA.

```python
r = release_dates
r = r[(r.country == 'USA') & (r.title.str.contains('Christmas'))].copy()
r['month'] = r.date.dt.month_name()

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

r = r.groupby(['month']).size().reindex(month_order)
r.plot(kind='bar', title='Number of Christmas movies released by month in USA')
```

### Make a bar plot of the months in which movies whose titles start with "The Hobbit" are released in the USA.

```python
r = release_dates
r = r[(r.country == 'USA') & (r.title.str.startswith('The Hobbit'))].copy()

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

r['month'] = r.date.dt.month_name()
r = r.groupby(['month']).size().reindex(month_order)
r.plot(kind='bar')
```

### Make a bar plot of the day of the week on which movies with "Romance" in their title tend to be released in the USA.

```python
r = release_dates
r = r[(r.country == 'USA') & (r.title.str.contains('Romance'))].copy()

r['day'] = r.date.dt.day_name()

day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

r = r.groupby('day').size().reindex(day_order)
r.plot(kind='bar',title='Romdance moviews released by day of the week')
```

### On which date was each Judi Dench movie from the 1990s released in the USA?

```python
r = release_dates
c = cast
c = c[c.name.str.contains('Judi Dench')]
df = c.merge(r, on=['title','year'])
df = df[(df.country == 'USA') & (df.year // 10 == 199)]
df.date.sort_values()
```

### In which months do films with Judi Dench tend to be released in the USA?

```python
r = release_dates
c = cast
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']

r = r[r.country == 'USA']
c = c[c.name == 'Judi Dench']

df = c.merge(r, on=['title','year'])
df['month'] = df.date.dt.month_name()
df.groupby('month').size().reindex(month_order).plot(kind='bar', title='films with Judi Dench released in USA by Month')
```

### In which months do films with Tom Cruise tend to be released in the USA?

```python
c = cast
r = release_dates
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']

c = c[c.name == 'Tom Cruise']
r = r[r.country == 'USA']
df = c.merge(r, on=['title','year'])

df['month'] = df.date.dt.month_name()

df.groupby('month').size().reindex(month_order).plot(kind='bar')

```
